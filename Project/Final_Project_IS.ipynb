```python
def generate_rsa_key():
    private_key = rsa.generate_private_key(
        public_exponent=65537,
        key_size=2048,
        backend=default_backend()
    )
    public_key = private_key.public_key()

    private_key_pem = private_key.private_bytes(
        encoding=serialization.Encoding.PEM,
        format=serialization.PrivateFormat.PKCS8,
        encryption_algorithm=serialization.NoEncryption()
    )
    public_key_pem = public_key.public_bytes(
        encoding=serialization.Encoding.PEM,
        format=serialization.PublicFormat.SubjectPublicKeyInfo
    )

    return private_key_pem, public_key_pem
```

```python
from cryptography.hazmat.primitives import hashes
from cryptography.hazmat.primitives.asymmetric import padding

def encrypt_file(file_path, public_key_pem):
    with open(file_path, 'rb') as f:
        plaintext = f.read()

    public_key = serialization.load_pem_public_key(public_key_pem, backend=default_backend())

    ciphertext = public_key.encrypt(
        plaintext,
        padding.OAEP(
            mgf=padding.MGF1(algorithm=hashes.SHA256()),
            algorithm=hashes.SHA256(),
            label=None
        )
    )

    return ciphertext
```

```python
def decrypt_file(encrypted_data, private_key_pem):
    private_key = serialization.load_pem_private_key(private_key_pem, password=None, backend=default_backend())

    decrypted_data = private_key.decrypt(
        encrypted_data,
        padding.OAEP(
            mgf=padding.MGF1(algorithm=hashes.SHA256()),
            algorithm=hashes.SHA256(),
            label=None
        )
    )

    return decrypted_data
```

```python
import hashlib

def generate_hash(file_path):
    hasher = hashlib.sha256()
    with open(file_path, 'rb') as f:
        for chunk in iter(lambda: f.read(4096), b""):
            hasher.update(chunk)
    return hasher.digest()
```

```python
def verify_integrity(received_file_path, original_hash):
    received_hash = generate_hash(received_file_path)
    return received_hash == original_hash
```

```python
import argparse
from cryptography.hazmat.primitives import serialization
from cryptography.hazmat.primitives.asymmetric import padding
from cryptography.hazmat.primitives import hashes
from cryptography.hazmat.backends import default_backend

def generate_rsa_key():
    private_key = serialization.load_pem_private_key(
        b"private_key", password=None, backend=default_backend()
    )
    public_key = serialization.load_pem_public_key(
        b"public_key", backend=default_backend()
    )
    return private_key, public_key

def encrypt_file(input_file, public_key_path, output_file):
    with open(public_key_path, 'rb') as f:
        public_key = serialization.load_pem_public_key(
            f.read(), backend=default_backend()
        )

    with open(input_file, 'rb') as f:
        plaintext = f.read()

    encrypted_data = public_key.encrypt(
        plaintext,
        padding.OAEP(
            mgf=padding.MGF1(algorithm=hashes.SHA256()),
            algorithm=hashes.SHA256(),
            label=None
        )
    )

    with open(output_file, 'wb') as f:
        f.write(encrypted_data)

def decrypt_file(input_file, private_key_path, output_file):
    with open(private_key_path, 'rb') as f:
        private_key = serialization.load_pem_private_key(
            f.read(), password=None, backend=default_backend()
        )

    with open(input_file, 'rb') as f:
        encrypted_data = f.read()

    decrypted_data = private_key.decrypt(
        encrypted_data,
        padding.OAEP(
            mgf=padding.MGF1(algorithm=hashes.SHA256()),
            algorithm=hashes.SHA256(),
            label=None
        )
    )

    with open(output_file, 'wb') as f:
        f.write(decrypted_data)

def generate_hash(input_file):
    hash_object = hashes.Hash(hashes.SHA256(), backend=default_backend())

    with open(input_file, 'rb') as f:
        while True:
            data = f.read(65536)
            if not data:
                break
            hash_object.update(data)

    return hash_object.finalize()

def verify_integrity(input_file, original_hash):
    current_hash = generate_hash(input_file)
    return current_hash == original_hash

if __name__ == '__main__':
    parser = argparse.ArgumentParser(description='Secure File Transfer System')
    parser.add_argument('mode', choices=['encrypt', 'decrypt', 'hash', 'verify'], help='Operation mode')
    parser.add_argument('--input', required=True, help='Input file path')
    parser.add_argument('--output', help='Output file path (for encryption/decryption)')
    parser.add_argument('--public_key', help='Path to recipient\'s public key file (for encryption)')
    parser.add_argument('--private_key', help='Path to your private key file (for decryption)')
    parser.add_argument('--original_hash', help='Original hash value (for integrity verification)')

    args = parser.parse_args()

    if args.mode == 'encrypt':
        if not args.output or not args.public_key:
            parser.error('--output and --public_key are required for encryption mode')
        encrypt_file(args.input, args.public_key, args.output)
        print(f'File encrypted and saved to {args.output}')
    elif args.mode == 'decrypt':
        if not args.output or not args.private_key:
            parser.error('--output and --private_key are required for decryption mode')
        decrypt_file(args.input, args.private_key, args.output)
        print(f'File decrypted and saved to {args.output}')
    elif args.mode == 'hash':
        hash_value = generate_hash(args.input)
        print(f'SHA-256 hash of {args.input}: {hash_value.hex()}')
    elif args.mode == 'verify':
        if not args.original_hash:
            parser.error('--original_hash is required for verify mode')
        is_integrity_verified = verify_integrity(args.input, bytes.fromhex(args.original_hash))
        if is_integrity_verified:
            print('Integrity verified: File has not been tampered with.')
        else:
            print('Integrity verification failed: File has been tampered with!')
    else:
        parser.error('Invalid mode selected')
```
